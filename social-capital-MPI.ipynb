```python
cols_2021 = [col for col in df.columns if '2021' in col]
cols_2021.append('Region')
```

```python
df_2021 = df[cols_2021]
```

```python
social_capita_df = df_2021.set_index('Region')
social_capita_df.head()
```

```python
social_capita_df.drop(['Italia'],axis=0,inplace=True)
```

```python
elementary_indicators = social_capita_df.columns
for column in elementary_indicators:
    social_capita_df[column] = 100+((social_capita_df[column]-social_capita_df[column].mean())/social_capita_df[column].std())*10
```

```python
social_capita_df['row_mean'] = social_capita_df[elementary_indicators].mean(axis=1)
social_capita_df['row_sd'] = social_capita_df[elementary_indicators].std(axis=1)
social_capita_df['cv'] = social_capita_df['row_sd']/social_capita_df['row_mean']
social_capita_df['penalty'] = social_capita_df['row_sd']*social_capita_df['cv']
social_capita_df['MPI'] = social_capita_df['row_mean']-social_capita_df['penalty']
social_capita_df['MPI_Rank'] = social_capita_df['MPI'].rank(ascending=False)
social_capita_df = social_capita_df.sort_values(by=['MPI_Rank'])
```

```python
social_capita_mpi = social_capita_df[['MPI']]
social_capita_mpi['DEN_REG'] = social_capita_mpi.index
social_capita_mpi
```
